```python
import pandas as pd 
```

```python
LDT_s1 = pd.read_csv('EZ_LDT_s1_correlation_matrix_renamed.csv')
LDT_s2 = pd.read_csv('EZ_LDT_s2_correlation_matrix_renamed.csv')
RMT_s1 = pd.read_csv('EZ_RMT_s1_correlation_matrix_renamed.csv')
RMT_s2 = pd.read_csv('EZ_RMT_s2_correlation_matrix_renamed.csv')
```

```python
sum=LDT_s1+LDT_s2+RMT_s1+RMT_s2
sum['Unnamed: 0'] = LDT_s1['Unnamed: 0']
ave=sum.loc[:,sum.columns != 'Unnamed: 0']/4
idx = 0 
ave.insert(loc=idx, column='par', value=LDT_s1['Unnamed: 0'])
```

```python
ave = ave.fillna(1)
```

```python
ave= ave.round(3)
```
