```python
import pandas as pd 
# Load the new dataset and display its first few rows to understand its structure
new_data_path = 'params_study2_retest_rt_acc_quartiles.csv'
new_data = pd.read_csv(new_data_path)

new_data.head()
```

```python
list(new_data)
```

```python
# Extract subject_id, task, and session from 'Experiment' column
new_data['subject_id'] = new_data['Experiment'].str.extract(r'(\d+)_').astype(int)
new_data['session'] = new_data['Experiment'].str.extract(r'_(part_\d)').astype(str)

# Now let's split the data into the four categories we are interested in
# For the Lexical Decision Task (LDT)
ldt_part_1_data_new = new_data[(new_data['session'] == 'part_1')]
ldt_part_2_data_new = new_data[(new_data['session'] == 'part_2')]



# Check if the splits are correctly done by displaying the head of each
(ldt_part_1_data_new.head(), ldt_part_2_data_new.head())
```

```python
# Investigate the unique values in the 'task' and 'session' columns to correct the filtering
unique_sessions = new_data['session'].unique()

unique_sessions
```

```python
# Correct parsing of the 'task' column to separate the 'class' and 'part' correctly
new_data['class'] = new_data['Experiment'].str.extract(r'class_(\w+)')

# Now let's split the data into the four categories we are interested in again
# For the Lexical Decision Task (LDT)
ldt_part_1_data_new = new_data[(new_data['class'] == 'words') & (new_data['session'] == 'part_1')]
ldt_part_2_data_new = new_data[(new_data['class'] == 'words') & (new_data['session'] == 'part_2')]

# For the Recognition Memory Task (RMT)
rmt_part_1_data_new = new_data[(new_data['class'] == 'pictures') & (new_data['session'] == 'part_1')]
rmt_part_2_data_new = new_data[(new_data['class'] == 'pictures') & (new_data['session'] == 'part_2')]

# Check if the splits are correctly done by displaying the head of each
(ldt_part_1_data_new.head(), ldt_part_2_data_new.head(), rmt_part_1_data_new.head(), rmt_part_2_data_new.head())
```
