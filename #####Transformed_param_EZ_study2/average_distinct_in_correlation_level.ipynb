```python
import pandas as pd 
```

```python
apt_s1 = pd.read_csv('EZ_Distinct_APT_s1_correlation_matrix.csv')
apt_s2 = pd.read_csv('EZ_Distinct_APT_s2_correlation_matrix.csv')
apt_s1
```

```python
sum=apt_s1+apt_s2
sum['Unnamed: 0'] = apt_s1['Unnamed: 0']
ave=sum.loc[:,sum.columns != 'Unnamed: 0']/2
idx = 0 
ave.insert(loc=idx, column='par', value=apt_s1['Unnamed: 0'])
```

```python
ave = ave.fillna(1)
ave= ave.round(3)

ave.to_csv('EZ_Study_2_average_distinct_correlation_across_2_data.csv')
```
