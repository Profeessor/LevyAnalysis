```python
import pandas as pd

# Load the new dataset
file_path_associative_learning = 'EZ_study2_params_plus_rt(log&main)_acc.csv'
associative_learning_data = pd.read_csv(file_path_associative_learning)

# Display the first few rows of the dataframe for a preview
associative_learning_data.head()
```

```python
associative_learning_data_part_1 = associative_learning_data[associative_learning_data['Experiment'].str.contains('part_1')]
associative_learning_data_part_2 = associative_learning_data[associative_learning_data['Experiment'].str.contains('part_2')]
```

```python
def process_dataframe(df):
    # Extracting participant's ID from the 'Experiment' column
    df['participant_ID'] = df['Experiment'].str.extract(r'(\d+)')
    # Converting ID to integer for proper sorting
    df['participant_ID'] = pd.to_numeric(df['participant_ID'])
    # Sorting the dataframe by the new ID column
    df_sorted = df.sort_values(by='participant_ID')
    return df_sorted
```

```python
from scipy import stats
import numpy as np

# Function to apply the transformations
def apply_transformations(df):
    # Transformation for alpha
    df['alpha_mean_shifted'] = df['alpha_mean'] - 1
    df['alpha_mean_arcsin'] = np.arcsin(np.sqrt(df['alpha_mean_shifted'].clip(0)))
    df['alpha_mean_boxcox_after_arcsin'], _ = stats.boxcox(df['alpha_mean_arcsin'] + 1e-9) # Adding a small constant to avoid zero values

    # Transformation for accuracy_percentage
    df['accuracy_percentage_arcsin'] = np.arcsin(np.sqrt(df['accuracy_percentage']))
    df['accuracy_percentage_boxcox_after_arcsin'], _ = stats.boxcox(df['accuracy_percentage_arcsin'] + 1e-9) # Adding a small constant to avoid zero values

    return df
```

```python
# Applying the ID extraction and sorting
associative_learning_data_part_1_sorted = process_dataframe(associative_learning_data_part_1)

# Applying the transformations
associative_learning_data_part_1_transformed = apply_transformations(associative_learning_data_part_1_sorted)

# Saving the transformed dataframe to a new CSV file
file_associative_learning_transformed_part_1 = 'EZ_associative_learning_transformed_part_1.csv'
associative_learning_data_part_1_transformed.to_csv(file_associative_learning_transformed_part_1, index=False)

file_associative_learning_transformed_part_1
```

```python
associative_learning_data_part_2_sorted = process_dataframe(associative_learning_data_part_2)

# Applying the transformations
associative_learning_data_part_2_transformed = apply_transformations(associative_learning_data_part_2_sorted)

# Saving the transformed dataframe to a new CSV file
file_associative_learning_transformed_part_2 = 'EZ_associative_learning_transformed_part_2.csv'
associative_learning_data_part_2_transformed.to_csv(file_associative_learning_transformed_part_2, index=False)

file_associative_learning_transformed_part_2
```
